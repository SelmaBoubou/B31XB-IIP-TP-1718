### 3.1  Resizing an image size using different interpolation functions

** Q.1 Resize the `lena` or `vibot` image using nearest, bilinear and bicubic function.** 

You can define your function which accept the model and resize size from the user or just run a script.

```python
# Load the lena image 
filepath = "./images/lena-grey.bmp"
Lena = io.imread(filepath)


# Conversion to float 
from skimage import img_as_float
Lena_im_float = img_as_float(Lena)

# size of the image 
size = 0.5 # to obtain 50% of the image

# Resize Lena image

Lb = misc.imresize(Lena, size, interp='bilinear')

Lc = misc.imresize(Lena, size, interp='bicubic')

Ln = misc.imresize(Lena, 0.2, interp='nearest')

# Dislay the resized images

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,20))
ax0, ax1, ax2, ax3 = axes.ravel()

ax0.imshow(Lb) #cmap = cm.grey
ax0.set_title('Resize image using bilinear method')

ax1.imshow(Lc)
ax1.set_title('Resize image using bicubic method')

ax2.imshow(Ln)
ax2.set_title('Resize image using nearest method')

ax3.imshow(Lena)
ax3.set_title('Original Image')
```
